```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear figura y ejes
fig, ax1 = plt.subplots()

# Generar histograma en el primer eje
ax1.hist(df['overall_rating'], bins=50, edgecolor='black', color='blue')
ax1.set_xlabel('??')
ax1.set_ylabel('overall_rating', color='blue')

# Crear un segundo eje y duplicar el primero
ax2 = ax1.twinx()

# Agregar la función de densidad (KDE) en el segundo eje
sns.kdeplot(df['overall_rating'], color='red', ax=ax2)
ax2.set_ylabel('Density (KDE)', color='red')

# Mostrar gráfico
plt.tight_layout()
plt.title('Histogram with Density')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar solo las variables continuas
columnas_continuas = df.select_dtypes(include='float64').columns

# Crear figura y ejes
plt.figure(figsize=(10, 6))

# Iterar sobre las columnas continuas y agregar la función de densidad
for columna in columnas_continuas:
    sns.kdeplot(df[columna], label=columna)

# Agregar leyenda y título
plt.legend()
plt.title('Density Plot for Continuous Variables')

# Configurar escala logarítmica en el eje x
plt.xscale('log')

# Mostrar gráfico
plt.show()


```
